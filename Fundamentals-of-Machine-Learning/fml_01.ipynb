## Select only a subset of columns to a new dataframe
* 'Id', 'SalePrice','MSSubClass','BldgType','HouseStyle','OverallQual','OverallCond','YearBuilt','Heating','CentralAir','GrLivArea','BedroomAbvGr'

```python
df = df.loc[:, ['Id', 'SalePrice','MSSubClass','BldgType','HouseStyle','OverallQual','OverallCond','YearBuilt','Heating','CentralAir','GrLivArea','BedroomAbvGr']].copy()
df.head()
```

### Select only houses built in year 2000 or later
* YearBuilt

```python
houses_2000_plus = df.loc[df['YearBuilt'] >= 2000]
houses_2000_plus.head()
```

### Select only houses that don't use GasA for heating (!= or ~ operators)
* Heating

```python
no_gas = df.loc[df['Heating'] != "GasA"]
```

### Select houses cheaper than 180k USD and with at least 2 bedrooms
* SalePrice, BedroomAbvGr

```python
cheap_houses = df.loc[(df['SalePrice'] < 180000) & (df['BedroomAbvGr'] >= 2)]
cheap_houses.head()
```

### Select houses with 2 stories or air conditioning
* HouseStyle, CentralAir

```python
two_stories_or_ac = df.loc[(df["HouseStyle"] == "2Story") | (df["CentralAir"] == "Y")]
two_stories_or_ac.head()
```

# Visualization

## Scatter plot
- Visualize relationship between SalePrice and GrLivArea.
- Use scatter plot from **Matplotlib**.

```python
plt.scatter(df["GrLivArea"], df["SalePrice"])
plt.show()
```

### Modify figure size and add title

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["GrLivArea"], df["SalePrice"])
plt.title("Relationship between SalePrice and GrLivArea")
plt.show()
```

### Add axis labels

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["GrLivArea"], df["SalePrice"])
plt.title("Relationship between SalePrice and GrLivArea")
plt.xlabel("GrLivArea (Above Ground Living Area)")
plt.ylabel("SalePrice (USD)")
plt.show()
```

### Add colors for data points based on CentralAir value.

```python
plt.figure(figsize=(8, 6))
colors = df["CentralAir"].map({"Y": "blue", "N": "red"})
plt.scatter(df["GrLivArea"], df["SalePrice"], c=colors)
plt.title("SalePrice vs GrLivArea (colored by CentralAir)")
plt.xlabel("GrLivArea (sq ft)")
plt.ylabel("SalePrice (USD)")
plt.show()
```

## Try to use scatterplot from **Seaborn** library for scatter plot visualization.

#### Use dataframe as source and column names for axes data

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="GrLivArea", y="SalePrice", hue="CentralAir")
plt.title("SalePrice vs GrLivArea (Seaborn)")
plt.show()
```

### Resize plot and add color for markers based on CentralAir column

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df,
    x="GrLivArea",
    y="SalePrice",
    hue="CentralAir",   # farby podƒæa CentralAir
    alpha=0.7           # priehƒæadnos≈• bodov
)
plt.title("SalePrice vs GrLivArea (colored by CentralAir)")
plt.xlabel("GrLivArea (sq ft)")
plt.ylabel("SalePrice (USD)")
plt.show()
```

## Line plot
- Calculate and vizualize average house price in relationship to YearBuild.

```python
avg_price_by_year = df.groupby("YearBuilt")["SalePrice"].mean()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(avg_price_by_year.index, avg_price_by_year.values, marker="o")
plt.title("Average Sale Price by Year Built")
plt.xlabel("Year Built")
plt.ylabel("Average Sale Price (USD)")
plt.grid(True)
plt.show()
```

## Bar plot
- Calculate and vizualize how many of houses have CentralAir
- Use Barplot for visualization

```python
central_air_counts = df["CentralAir"].value_counts()
```

```python
plt.figure(figsize=(6, 5))
sns.barplot(x=central_air_counts.index, y=central_air_counts.values)
plt.title("Count of Houses with and without Central Air")
plt.xlabel("CentralAir (Y/N)")
plt.ylabel("Number of Houses")
plt.show()
```

### Vizualize number of building type and if they contains air conditioning using Seaborn

```python
plt.figure(figsize=(9, 6))
sns.countplot(data=df, x="BldgType", hue="CentralAir")
plt.title("Building Type vs Central Air")
plt.xlabel("Building Type")
plt.ylabel("Number of Houses")
plt.show()
```
